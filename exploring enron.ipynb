```python


%matplotlib inline
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from operator import itemgetter
from IPython.core.display import HTML
pd.set_option('display.max_rows', 5000)

```

```python


enron_data = pickle.load(open("../ud120-projects/final_project/final_project_dataset.pkl", "r"))

```

```python
def print_quiz_answer(str):
    return HTML('<div style="color:#0000FF">QUIZ Anwer:  </dic><div style="color:#FF00FF">{}</dic>'.format(str))
```

```python
enron_df = pd.DataFrame(enron_data)
enron_df_t = enron_df.transpose()
```

```python
print_quiz_answer('number of persons: {}'.format(len(enron_df_t)))
```

```python
print_quiz_answer('number of features: {}'.format(len(enron_df_t.columns)))
```

```python
s = len(enron_df_t[enron_df_t['poi']])
print_quiz_answer('number of persons of interest: {}'.format(s))
```

```python
# read text files with names
fname = '../ud120-projects/final_project/poi_names.txt'
with open(fname) as f:
    content = f.readlines()

#skip two first lines
content = content[2:]

#find the values
persons = []

for lines in content:
    m = re.match('\((?P<_0>.+)\) (?P<_1>.+)\, (?P<_2>.+)\n', lines)
    persons.append(map(itemgetter(1), sorted(m.groupdict().items())))

s = len(persons)
print_quiz_answer('number of persons of interest in text file: {}'.format(s))
```

```python
cxo = []
for row_index, row in enron_df_t.iterrows():
    if row_index in major:
        #print('%s\n%s' % (row_index, row.total_payments))
        cxo.append([row_index, row.total_payments])

cxo = pd.DataFrame(cxo)
cxo = cxo.sort_values(by=1,ascending=False)
cxo[0:1][0]
print_quiz_answer('top earner: {}, with {}'.format(cxo[0:1][0],cxo[0:1][1]))
```
